```python
import numpy as np
def mylog(x):
    n = len(x)
    lnx = np.zeros(n)
    for i in range(n):
        if x[i] == 0:
            lnx[i] = 0
        else:
            lnx[i] = np.log(x[i])
    return lnx
n = int(input('请输入参评数目：'))
m = int(input('请输入指标数目：'))
X = np.zeros(shape = (n,m))
for i in range(n):
    X[i] = input().split(" ")
    X[i] = list(map(float,X[i])) 
Z = X/np.sqrt(np.sum(X*X,axis=0)) #标准化，使不同指标具有可比性
print(Z)

n,m = Z.shape
D = np.zeros(m)

for i in range(m):
    x = Z[:,i]
    X = x/np.sum(x)
    e = -np.sum(X*mylog(X))/np.log(n)
    D[i] = 1-e
W = D/np.sum(D)

print(f"权重W= {W}")
```

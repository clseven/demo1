## 非线性规划
示例：


```python
import numpy as np
from scipy.optimize import minimize

#目标函数
def fun(x):
    return x[0]**2+x[1]**2-x[0]*x[1]-2*x[0]-5*x[1]

#约束条件
def cons1(x):
    return -(x[0]-1)**2+x[1]
def cons2(x):
    return 2*x[0]-3*x[1]+6
cons = (
    {'type': 'ineq', 'fun': cons1},
    {'type': 'ineq', 'fun': cons2}
)
#边界约束
bounds = np.array([[0,None],[0,None]])
#猜测解
x0 = np.array([2.9925508333192425,3.851230384422948])

res = minimize(fun,x0,method='SLSQP', bounds=bounds, constraints=cons)
print(res)
```

```python
import numpy as np
# 可用蒙特卡洛法确定猜测解
n = 1000000
fmin = float('inf')
for i in range(n):
    x1 = np.random.uniform(-100,100)
    x2 = np.random.uniform(-100,100)
    if (-(x1-1)**2+x2**2)>=0 and 2*x1-3*x2+6>=0:
        result = x1**2+x2**2-x1*x2-2*x1-5*x2
        if result < fmin:
            fmin = result
            x1min = x1
            x2min = x2
print(fmin)
print(x1min,x2min)
```

# 主成分分析法（PCA）


```python
import pandas as pd
import numpy as np
import scipy

dr = pd.read_excel(r"C:\Users\16619\Downloads\棉花产量论文作业的数据.xlsx",usecols='C:G')

x = dr.to_numpy()
#标准化
X = (x - np.mean(x,axis = 0))/np.std(x,ddof = 1,axis = 0)
#协方差
R = np.cov(X.T)

#特征值，特征向量
eigva,eigve = scipy.linalg.eigh(R)#会按从小到大排列
eigva = eigva[::-1]
eigve = eigve[:,::-1]

rate = eigva/sum(eigva)
ratesum = np.cumsum(rate)

print('特征值为: ',eigva)
print('特征向量为: \n',eigve)
print('贡献率为: ',rate)
print('累计贡献率为: ',ratesum)
```

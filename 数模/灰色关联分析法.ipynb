```python
import numpy as np

#[[55,24,10],[65,38,22],[75,40,18],[100,50,20]]
# n = int(input('请输入参评数目：'))
# m = int(input('请输入指标数目：'))
# X = np.zeros(shape = (n,m))
# for i in range(n):
#     X[i] = input().split(" ")
#     X[i] = list(map(float,X[i])) 

X = np.array(eval(input()))#输入形式为列表

Mean = np.mean(X,axis=0)#求平均值

#数据预处理
Xa  = X/Mean
print(f"预处理后的矩阵：\n{Xa}")

M = Xa[:,0]
Z = Xa[:,1:]
absX0_Xi = np.abs(Z - np.tile(M.reshape(-1,1),(1,Z.shape[1])))

a = np.min(absX0_Xi)
b = np.max(absX0_Xi)

#分辨系数
rho = 0.5

#关联系数
Xi = (a+rho*b)/(absX0_Xi+rho*b)

print(f"子序列灰色关联度为：\n{np.mean(Xi,axis=0)}") 
```

```python
#评价类
#先正向化
#构造母序列子序列
#取每一行最大值构成母序列
weight = np.mean(Xi,axis=0)/np.sum(np.mean(Xi,axis=0))#通过灰色关联度计算权重
#计算得分且归一化
```

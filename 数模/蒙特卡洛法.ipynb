```python
#
import matplotlib.pyplot as plt
import numpy as np
from pylab import mpl
# 
mpl.rcParams["font.sans-serif"] = ["SimHei"]

p = 15000
r = 1;
x0,y0=1,1
n = 0#

plt.figure()
plt.title("")
plt.xlabel('x')
plt.ylabel('y')

for i in range(p):
    np.random.rand()
    px = np.random.rand()*2
    py = np.random.rand()*2
    
    if(px - x0)**2+(py - y0)**2<r:
        plt.plot(px,py,'.',color='b')
        n +=1
    else:
        plt.plot(px,py,'.',color='r')
plt.axis('equal')
plt.show()

pi = (n/p)*4
print(": ",pi)
```

```python
#
import numpy as np
a = 0 #
b = 0 #
c = 0 #
n = 100000

for i in range(n):
    tru = np.random.randint(1,4)
    me = np.random.randint(1,4)
    change = np.random.randint(0,2)
    if tru == me:
        if change == 0:
            a+=1
        else:
            c+=1
    else:
        if change == 1:
            b+=1
        else:
            c+=1
            
print('',a/n)
print('',b/n)
print('',c/n)
```

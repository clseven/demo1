# 蒙特卡洛法

```python
#圆周率问题
import matplotlib.pyplot as plt
import numpy as np
from pylab import mpl
# 设置显示中文字体
mpl.rcParams["font.sans-serif"] = ["SimHei"]

p = 15000
r = 1;
x0,y0=1,1
n = 0#圆内的点数

plt.figure()
plt.title("圆周率问题")
plt.xlabel('x')
plt.ylabel('y')

for i in range(p):
    np.random.rand()
    px = np.random.rand()*2
    py = np.random.rand()*2
    
    if(px - x0)**2+(py - y0)**2<r:
        plt.plot(px,py,'.',color='b')
        n +=1
    else:
        plt.plot(px,py,'.',color='r')
plt.axis('equal')
plt.show()

pi = (n/p)*4
print("估算的Π等于: ",pi)
```

```python
#三门问题
import numpy as np
a = 0 #不改变主意拿汽车的次数
b = 0 #改变主意拿汽车的次数
c = 0 #失败的次数
n = 100000

for i in range(n):
    tru = np.random.randint(1,4)
    me = np.random.randint(1,4)
    change = np.random.randint(0,2)
    if tru == me:
        if change == 0:
            a+=1
        else:
            c+=1
    else:
        if change == 1:
            b+=1
        else:
            c+=1
            
print('模拟后不改变主意的获奖概率为：',a/n)
print('模拟后改变主意的获奖概率为：',b/n)
print('失败的概率为：',c/n)
```

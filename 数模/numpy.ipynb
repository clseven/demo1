```python
#nan no a number
import numpy as np

t2=np.arange(0,24,1,dtype="float").reshape(4,6)
print(t2)
print("*"*100)

print(t2.sum(axis=0))
print("*"*100)
print(t2.sum(axis=1))
print("*"*100)

t2[t2>20]=20
t2[3,3]=np.nan
t2[3,4]=np.nan

a = np.inf
b = np.nan
print(a,b)

b!=b #True

print(t2)
print("*"*100)

#用nan!=nan 来判断nan的个数
np.count_nonzero(t2) #23
np.count_nonzero(t2!=t2)


for i in range (t2.shape[1]):
    t2now = t2[:,i]
    nan_num = np.count_nonzero(t2now!=t2now)
    if nan_num!=0:
        t2now_notnan = t2now[t2now==t2now]
        t2now[np.isnan(t2now)] = t2now_notnan.mean() #将nan赋值为该列均值
  
print(t2)
```

```python
#数组的拼接
import numpy as np

t1 = np.array([[0,1],[2,3]])
t2 = np.array([[4,5],[6,7]])

print(np.vstack((t1,t2))) #竖直拼接
print(np.hstack((t1,t2))) #水平拼接

#数组行列交换

t1[[0,1],:] = t1[[1,0],:] #行交换
print(t1)
t1[:,[0,1]] = t1[:,[1,0]] #列交换
print(t1)
```

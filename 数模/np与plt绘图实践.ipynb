```python
#结合np和plt 绘制评论数量直方图
import numpy as np
from matplotlib import pyplot as plt

usfile = "./US_video_data_numbers.csv"
ukfile = "./GB_video_data_numbers.csv"

tus = np.loadtxt(usfile,delimiter=",",dtype=int)
tuk = np.loadtxt(ukfile,delimiter=",",dtype=int)
                 
#取评论的数据(最后一列)
tus_comment = tus[:,-1]

tus_comment = tus_comment[tus_comment<=5000]



d = 100

bin_nums = (tus_comment.max()-tus_comment.min())//d

print(bin_nums)

plt.figure(figsize=(20,8),dpi=80) 

plt.hist(tus_comment,bin_nums) #对不齐可以传列表

plt.grid()

plt.show()
```

```python
#绘制评论数与喜欢数的关系
import numpy as np
from matplotlib import pyplot as plt

plt.rcParams['font.sans-serif'] = ['SimHei']  # 用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False  # 用来正常显示负号

usfile = "./US_video_data_numbers.csv"
ukfile = "./GB_video_data_numbers.csv"

tus = np.loadtxt(usfile,delimiter=",",dtype=int)
tuk = np.loadtxt(ukfile,delimiter=",",dtype=int)

tuk = tuk [tuk[:,1]<=50000]

tuk_comment = tuk[:,-1]
tuk_like = tuk[:,1]

plt.figure(figsize=(20,8),dpi=80)
plt.scatter(tuk_like,tuk_comment)

plt.xlabel("喜欢数")
plt.ylabel("评论数")
plt.title("喜欢数与评论数的关系图")

plt.show()
```

```python
import numpy as np
from matplotlib import pyplot as plt

plt.rcParams['font.sans-serif'] = ['SimHei']  # 用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False  # 用来正常显示负号

usfile = "./US_video_data_numbers.csv"
ukfile = "./GB_video_data_numbers.csv"

tus = np.loadtxt(usfile,delimiter=",",dtype=int)
tuk = np.loadtxt(ukfile,delimiter=",",dtype=int)

zeros_data = np.zeros((tus.shape[0],1)).astype(int)
ones_data = np.ones((tuk.shape[0],1)).astype(int)

tus = np.hstack((tus,zeros_data))
tuk = np.hstack((tuk,ones_data))

final_data = np.vstack((tus,tuk))
print(final_data)
```

```python
def floyd(g):
    n = len(g)
    for k in range(n):
        for i in range(n):
            for j in range(n):
                if (g[i][j]>g[i][k]+g[k][j] or g[i][j]==-1)and g[i][k]>0 and g[k][j]>0:
                    g[i][j] = g[i][k]+g[k][j]
    for row in g:
        print(row) #每一个点到其他点的最短路径
matrix = [[0, -1, -1, 1.2, 9.1, -1, 0.5],
          [-1, 0, -1, 5, -1, 3.1, 2],
          [-1, -1, 0, -1, -1, 4, 1.5],
          [1.2, 5, -1, 0, 6.7, -1, -1],
          [9.2, -1, -1, 6.7, 0, 15.6, -1],
          [-1, 3.1, 4, -1, 15.6, 0, -1],
          [0.5, 2, 1.5, -1, -1, -1, 0]]
floyd(matrix)
```
